```python

import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
def latex_plot(scale = 1, fontsize = 12):
    """ Changes the size of a figure and fonts for the publication-quality plots. """
    fig_width_pt = 246.0
    inches_per_pt = 1.0/72.27
    golden_mean = (np.sqrt(5.0)-1.0)/2.0
    fig_width = fig_width_pt*inches_per_pt*scale
    fig_height = fig_width*golden_mean
    fig_size = [fig_width, fig_height]
    eps_with_latex = {
        "pgf.texsystem": "pdflatex", "text.usetex": True, "font.family": "serif", "font.serif": [], "font.sans-serif": [], "font.monospace": [], "axes.labelsize": fontsize, "font.size": fontsize, "legend.fontsize": fontsize, "xtick.labelsize": 12, "ytick.labelsize": 12, "figure.figsize": fig_size
        }
    mpl.rcParams.update(eps_with_latex)
```

```python
# Watts-Strogatz graph
ws_k2 = np.loadtxt('ws_robustness_N_10000_L_100_beta_0.010_k_2.dat')
f_ws = ws_k2[:, 0]
random_ws_k2 = ws_k2[:, 1]/ws_k2[0,1]
degree_ws_k2 = ws_k2[:, 2]/ws_k2[0,2]
betweenness_ws_k2 = ws_k2[:, 3]/ws_k2[0,3]
closeness_ws_k2 = ws_k2[:, 4]/ws_k2[0,4]
ws_k4 = np.loadtxt('ws_robustness_N_10000_L_100_beta_0.010_k_4.dat')
random_ws_k4 = ws_k4[:, 1]/ws_k4[0,1]
degree_ws_k4 = ws_k4[:, 2]/ws_k4[0,2]
betweenness_ws_k4 = ws_k4[:, 3]/ws_k4[0,3]
closeness_ws_k4 = ws_k4[:, 4]/ws_k4[0,4]
```

# Robustness against random attacks

```python
# grid of 2x2 subplots using gridspec
latex_plot(scale=1.5, fontsize=16)
fig = plt.figure(figsize=(18,12))
gs = fig.add_gridspec(2, 2)
ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[1, 0])
ax4 = fig.add_subplot(gs[1, 1])

plt.suptitle("Robustness against random attacks", fontsize=22)

ax1.set_title(r'Average degree $\langle k \rangle = 0.5$', fontsize=16)
ax2.set_title(r'Average degree $\langle k \rangle = 1$', fontsize=16)
ax3.set_title(r'Average degree $\langle k \rangle = 2$', fontsize=16)
ax4.set_title(r'Average degree $\langle k \rangle = 4$', fontsize=16)

ax1.set_ylabel(r"Size of giant component $P_{\infty}(f)/P_{\infty}(0)$", fontsize=16)
ax3.set_ylabel(r"Size of giant component $P_{\infty}(f)/P_{\infty}(0)$", fontsize=16)
ax3.set_xlabel(r"Fraction of removed nodes $f$", fontsize=16)
ax4.set_xlabel(r"Fraction of removed nodes $f$", fontsize=16)

# tick labels size
ax1.tick_params(axis='both', which='major', labelsize=16)
ax2.tick_params(axis='both', which='major', labelsize=16)
ax3.tick_params(axis='both', which='major', labelsize=16)
ax4.tick_params(axis='both', which='major', labelsize=16)


# ws graph
ax3.plot(f_ws, random_ws_k2, label="Watts-Strogatz", marker='o', linestyle='--')
ax4.plot(f_ws, random_ws_k4, label="Watts-Strogatz", marker='o', linestyle='--')

plt.legend(loc='best', fontsize=16)
plt.tight_layout()
    
```
