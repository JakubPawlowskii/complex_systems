```python

import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
def latex_plot(scale = 1, fontsize = 12):
    """ Changes the size of a figure and fonts for the publication-quality plots. """
    fig_width_pt = 246.0
    inches_per_pt = 1.0/72.27
    golden_mean = (np.sqrt(5.0)-1.0)/2.0
    fig_width = fig_width_pt*inches_per_pt*scale
    fig_height = fig_width*golden_mean
    fig_size = [fig_width, fig_height]
    eps_with_latex = {
        "pgf.texsystem": "pdflatex", "text.usetex": True, "font.family": "serif", "font.serif": [], "font.sans-serif": [], "font.monospace": [], "axes.labelsize": fontsize, "font.size": fontsize, "legend.fontsize": fontsize, "xtick.labelsize": 12, "ytick.labelsize": 12, "figure.figsize": fig_size
        }
    mpl.rcParams.update(eps_with_latex)
```

```python
# Watts-Strogatz graph
ws_k2 = np.loadtxt('100_ws_robustness_N_10000_L_100_beta_0.010_k_2.dat')
f_ws = ws_k2[:, 0]
random_ws_k2 = ws_k2[:, 1]/ws_k2[0,1]
degree_ws_k2 = ws_k2[:, 2]/ws_k2[0,2]
betweenness_ws_k2 = ws_k2[:, 3]/ws_k2[0,3]
closeness_ws_k2 = ws_k2[:, 4]/ws_k2[0,4]
ws_k4 = np.loadtxt('100_ws_robustness_N_10000_L_100_beta_0.010_k_4.dat')
random_ws_k4 = ws_k4[:, 1]/ws_k4[0,1]
degree_ws_k4 = ws_k4[:, 2]/ws_k4[0,2]
betweenness_ws_k4 = ws_k4[:, 3]/ws_k4[0,3]
closeness_ws_k4 = ws_k4[:, 4]/ws_k4[0,4]
```

```python
# Barabasi-Albert graph
# ba_k1 = np.loadtxt('100_ba_robustness_N_10000_L_100_k_1.dat')
# f_ba = ba_k1[:, 0]
# random_ba_k1 = ba_k1[:, 1]/ba_k1[0,1]
# degree_ba_k1 = ba_k1[:, 2]/ba_k1[0,2]
# betweenness_ba_k1 = ba_k1[:, 3]/ba_k1[0,3]
# closeness_ba_k1 = ba_k1[:, 4]/ba_k1[0,4]

ba_k2 = np.loadtxt('100_ba_robustness_N_10000_L_100_k_2.dat')
f_ba = ba_k2[:, 0]
random_ba_k2 = ba_k2[:, 1]/ba_k2[0,1]
degree_ba_k2 = ba_k2[:, 2]/ba_k2[0,2]
betweenness_ba_k2 = ba_k2[:, 3]/ba_k2[0,3]
closeness_ba_k2 = ba_k2[:, 4]/ba_k2[0,4]
ba_k4 = np.loadtxt('100_ba_robustness_N_10000_L_100_k_4.dat')
random_ba_k4 = ba_k4[:, 1]/ba_k4[0,1]
degree_ba_k4 = ba_k4[:, 2]/ba_k4[0,2]
betweenness_ba_k4 = ba_k4[:, 3]/ba_k4[0,3]
closeness_ba_k4 = ba_k4[:, 4]/ba_k4[0,4]
```

```python
# Erdos-Renyi graph
er_k05 = np.loadtxt('100_er_robustness_N_10000_L_100_k_0.5.dat')
f_er = er_k05[:, 0]
random_er_k05 = er_k05[:, 1]/er_k05[0,1]
degree_er_k05 = er_k05[:, 2]/er_k05[0,2]
betweenness_er_k05 = er_k05[:, 3]/er_k05[0,3]
closeness_er_k05 = er_k05[:, 4]/er_k05[0,4]
er_k1 = np.loadtxt('100_er_robustness_N_10000_L_100_k_1.dat')
random_er_k1 = er_k1[:, 1]/er_k1[0,1]
degree_er_k1 = er_k1[:, 2]/er_k1[0,2]
betweenness_er_k1 = er_k1[:, 3]/er_k1[0,3]
closeness_er_k1 = er_k1[:, 4]/er_k1[0,4]
er_k2 = np.loadtxt('100_er_robustness_N_10000_L_100_k_2.dat')
random_er_k2 = er_k2[:, 1]/er_k2[0,1]
degree_er_k2 = er_k2[:, 2]/er_k2[0,2]
betweenness_er_k2 = er_k2[:, 3]/er_k2[0,3]
closeness_er_k2 = er_k2[:, 4]/er_k2[0,4]

er_k4 = np.loadtxt('100_er_robustness_N_10000_L_100_k_4.dat')
random_er_k4 = er_k4[:, 1]/er_k4[0,1]
degree_er_k4 = er_k4[:, 2]/er_k4[0,2]
betweenness_er_k4 = er_k4[:, 3]/er_k4[0,3]
closeness_er_k4 = er_k4[:, 4]/er_k4[0,4]
```

# Robustness against random attacks

```python
# grid of 2x2 subplots using gridspec
latex_plot(scale=1.5, fontsize=16)
fig = plt.figure(figsize=(18,12))
gs = fig.add_gridspec(2, 2)
ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[1, 0])
ax4 = fig.add_subplot(gs[1, 1])

plt.suptitle("Robustness against random attacks, $N=10^4,\; L = 100$", fontsize=22)

ax1.set_title(r'Average degree $\langle k \rangle = 0.5$', fontsize=16)
ax2.set_title(r'Average degree $\langle k \rangle = 1$', fontsize=16)
ax3.set_title(r'Average degree $\langle k \rangle = 2$', fontsize=16)
ax4.set_title(r'Average degree $\langle k \rangle = 4$', fontsize=16)

ax1.set_ylabel(r"Size of giant component $P_{\infty}(f)/P_{\infty}(0)$", fontsize=16)
ax3.set_ylabel(r"Size of giant component $P_{\infty}(f)/P_{\infty}(0)$", fontsize=16)
ax3.set_xlabel(r"Fraction of removed nodes $f$", fontsize=16)
ax4.set_xlabel(r"Fraction of removed nodes $f$", fontsize=16)

# tick labels size
ax1.tick_params(axis='both', which='major', labelsize=16)
ax2.tick_params(axis='both', which='major', labelsize=16)
ax3.tick_params(axis='both', which='major', labelsize=16)
ax4.tick_params(axis='both', which='major', labelsize=16)


ax1.plot(f_er, random_er_k05, label="Erdos-Renyi", marker='^', linestyle='--')

ax2.plot(f_er, random_er_k1, label="Erdos-Renyi", marker='^', linestyle='--')
# ax2.plot(f_ba, random_ba_k1, label="Barabasi-Albert", marker='s', linestyle='--')
ax2.legend(loc="upper right", fontsize=16)


ax3.plot(f_er, random_er_k2, label="Erdos-Renyi", marker='^', linestyle='--')
ax3.plot(f_ba, random_ba_k2, label="Barabasi-Albert", marker='s', linestyle='--')
ax3.plot(f_ws, random_ws_k2, label=r"Watts-Strogatz, $\beta=0.01$", marker='o', linestyle='--')


ax4.plot(f_er, random_er_k4, label="Erdos-Renyi", marker='^', linestyle='--')
ax4.plot(f_ba, random_ba_k4, label="Barabasi-Albert", marker='s', linestyle='--')
ax4.plot(f_ws, random_ws_k4, label=r"Watts-Strogatz, $\beta=0.01$", marker='o', linestyle='--')

ax1.legend(loc="upper right", fontsize=16)
ax3.legend(loc="upper right", fontsize=16)
ax4.legend(loc="upper right", fontsize=16)


# caption below the figure

plt.figtext(0.5, -0.1, "Figure 1: Robustness against random attacks for different graphs. The fraction of removed nodes $f$ is varied from 0 to 1.\
 The size of the giant component\n $P_{\infty}(f)$ is normalized by the size of the giant component $P_{\infty}(0)$ for the original graph.\
    Watts-Strogatz graph is the most vulnerable one, because for $\\beta = 0.01$,\n it is in the small world regime, and it is easy for the components to be disconnected.\
    Note the suprising resilience of Erdos-Renyi graph for $ \\langle k \\rangle = 0.5$.\n This is due to many disconnected components exisitng for such small edge probability.", ha="center", fontsize=20)

plt.tight_layout()
    
```
